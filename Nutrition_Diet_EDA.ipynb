```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df=pd.read_csv("C:/Users/dell/Downloads/All_Diets.csv")
df.info()
df.dtypes
```

```python
data=df.drop(['Extraction_day','Extraction_time'],axis=1)
         
                                  
data.rename(columns={'Protein(g)':'Protein','Carbs(g)':'Carbs','Fat(g)':'Fat'},inplace=True)

data['Calories'] = data['Protein']*4 + data['Carbs']*4 + data['Fat']*9
```

```python
#High Protein/Low Carb - different cuisine

data['Cuisine_type'].unique()
#'american', 'south east asian', 'mexican', 'chinese',
#'mediterranean', 'italian', 'french', 'indian', 'nordic',
#'eastern europe', 'central europe', 'kosher', 'british',
#'caribbean', 'south american', 'middle eastern', 'asian',
#'japanese', 'world'
#Will select only five

dt_american=data.loc[(data['Protein']/3>data['Carbs']) & (data['Cuisine_type']=='american')]
dt_american.head(5)

dt_Italian=data.loc[(data['Protein']/3>data['Carbs']) & (data['Cuisine_type']=='italian')]
dt_Italian.head(5)


dt_mediterranean=data.loc[(data['Protein']/3>data['Carbs']) & (data['Cuisine_type']=='mediterranean')]
dt_mediterranean.head(5)

dt_french=data.loc[(data['Protein']/3>data['Carbs']) & (data['Cuisine_type']=='french')]
dt_french.head(5)

dt_indian=data.loc[(data['Protein']/3>data['Carbs']) & (data['Cuisine_type']=='indian')]
dt_indian.head(5)
```

```python
#High Protein Food - Less Fat
df_american=data.loc[(data['Protein']/3>data['Fat']) & (data['Cuisine_type']=='american')]
df_american.head(5)

df_Italian=data.loc[(data['Protein']/3>data['Fat']) & (data['Cuisine_type']=='italian')]
df_Italian.head(5)


df_mediterranean=data.loc[(data['Protein']/3>data['Fat']) & (data['Cuisine_type']=='mediterranean')]
df_mediterranean.head(5)

df_french=data.loc[(data['Protein']/3>data['Fat']) & (data['Cuisine_type']=='french')]
df_french.head(5)

df_indian=data.loc[(data['Protein']/3>data['Fat']) & (data['Cuisine_type']=='indian')]
df_indian.head(5)
```

```python
#High Protein - Less Fat, less Carbs

#american
america_rich_protein=data.loc[(data['Protein']/3>data['Fat']) & 
                     (data['Protein']/3>data['Carbs']) &
                     (data['Cuisine_type']=='american')]
america_rich_protein.head(5)

#indian
indian_rich_protein=data.loc[(data['Protein']/3>data['Fat']) & 
                     (data['Protein']/3>data['Carbs']) &
                     (data['Cuisine_type']=='indian')]
indian_rich_protein.head(5)

#italian
italian_rich_protein=data.loc[(data['Protein']/3>data['Fat']) & 
                     (data['Protein']/3>data['Carbs']) &
                     (data['Cuisine_type']=='italian')]
italian_rich_protein.head(5)


#mediterranean
mediterranean_rich_protein=data.loc[(data['Protein']/3>data['Fat']) & 
                     (data['Protein']/3>data['Carbs']) &
                     (data['Cuisine_type']=='mediterranean')]
mediterranean_rich_protein.head(5)

#french
french_rich_protein=data.loc[(data['Protein']/3>data['Fat']) & 
                     (data['Protein']/3>data['Carbs']) &
                     (data['Cuisine_type']=='french')]
french_rich_protein.head(5)
```

```python
# Unhealthiest_food
#American
unhealthy_food_american=data.loc[(data['Carbs']*4 >=(data['Calories']/100)*45) & 
                     (data['Fat']*9>=(data['Calories']/100)*35) &
                     (data['Cuisine_type']=='american')]
unhealthy_food_american.head(10)

#Indian
unhealthy_food_indian=data.loc[(data['Carbs']*4 >=(data['Calories']/100)*45) & 
                     (data['Fat']*9>=(data['Calories']/100)*35) &
                     (data['Cuisine_type']=='indian')]
unhealthy_food_indian.head(10)

#Italian
unhealthy_food_italian=data.loc[(data['Carbs']*4 >=(data['Calories']/100)*45) & 
                     (data['Fat']*9>=(data['Calories']/100)*35) &
                     (data['Cuisine_type']=='italian')]
unhealthy_food_italian.head(10)


#mediterranean
unhealthy_food_mediterranean=data.loc[(data['Carbs']*4 >=(data['Calories']/100)*45) & 
                     (data['Fat']*9>=(data['Calories']/100)*35) &
                     (data['Cuisine_type']=='mediterranean')]
unhealthy_food_mediterranean.head(10)

#french
unhealthy_food_french=data.loc[(data['Carbs']*4 >=(data['Calories']/100)*45) & 
                     (data['Fat']*9>=(data['Calories']/100)*35) &
                     (data['Cuisine_type']=='french')]
unhealthy_food_french.head(10)
```

```python

unhealthy_food=data.loc[(data['Carbs']*4 >=(data['Calories']/100)*45) & 
                     (data['Fat']*9>=(data['Calories']/100)*35)].sort_values(by='Calories',ascending=False)
unhealthy_food.head(10)
```

```python
fig = plt.figure(figsize = (10, 5))

plt.title('Unhealthiest Food In Different Cuisines')
sn.set(style="ticks")
sn.countplot(y='Cuisine_type',data=unhealthy_food,
             order =unhealthy_food['Cuisine_type'].value_counts().index,
             palette="rocket_d")
sn.set_style("darkgrid")
plt.xlabel('Counts')
plt.ylabel('Cuisine Type')
plt.show()
```

```python
fig = plt.figure(figsize = (10, 5))

plt.title('Unhealthiest Food In Different Diet Type')
sn.set(style="ticks")
sn.countplot(y='Diet_type',data=unhealthy_food,
             order =unhealthy_food['Diet_type'].value_counts().index,
             palette="magma",ec='Black')
sn.set_style("darkgrid")
plt.xlabel('Counts')
plt.ylabel('Diet Type')
plt.show()
```
